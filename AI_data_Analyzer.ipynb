```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('ggplot')

print("Libraries loaded successfully.")
```

```python
data = {
    'StudentID': np.arange(101, 111),
    'Gender': ['F', 'M', 'F', 'M', 'F', 'M', 'F', 'M', 'F', 'M'],
    'Hours_Studied': [5, 8, 3, 10, 6, 4, 7, 9, 2, 5],
    'Exam_Score': [85, 92, 78, 95, 88, 75, 90, 98, 70, 80],
    'Attendance_Rate': [95.0, 100.0, 90.0, 100.0, 98.0, 85.0, 99.0, 96.0, 92.0, 97.0]
}
df = pd.DataFrame(data)
file_name = 'student_performance.csv'
df.to_csv(file_name, index=False)
```

```python
df_loaded = pd.read_csv('student_performance.csv')
print("First 5 rows of the loaded dataset:")
print(df_loaded.head())

print("\nData information (columns and data types):")
print(df_loaded.info())
```

```python
print("\nMissing values per column:")
print(df_loaded.isnull().sum())
```

```python
print("\n--- Summary Statistics (Exam_Score, Hours_Studied) ---")
print(df_loaded[['Exam_Score', 'Hours_Studied']].describe())

mean_score_by_gender = df_loaded.groupby('Gender')['Exam_Score'].mean().reset_index()
print("\n--- Mean Exam Score by Gender ---")
print(mean_score_by_gender)

correlation = df_loaded['Hours_Studied'].corr(df_loaded['Exam_Score'])
print(f"\nCorrelation between Hours Studied and Exam Score: {correlation:.4f}")
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df_loaded['Hours_Studied'], df_loaded['Exam_Score'], color='blue', alpha=0.7)
plt.title('Relationship between Hours Studied and Exam Score')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.grid(True)
plt.savefig('hours_vs_score_scatter.png')

print("Scatter plot generated and saved as 'hours_vs_score_scatter.png'.")
```
