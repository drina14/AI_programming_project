```python
import re
import sys
response_data = {
    ('hello', 'hi', 'hey', 'start'):
        "Hello! I am your Grade Predictor Assistant Bot. I can help answer common questions about calculating and improving your academic score. How can I assist you today?",
    ('factor', 'influence', 'determine', 'affect', 'what counts'):
        "Great question! Your overall grade is typically influenced by three main factors: **1) Assignments and Quizzes**, **2) Midterm and Final Exams**, and **3) Participation or Attendance**. Always check your course syllabus for exact weightings!",
    ('calculate', 'formula', 'weight', 'final grade', 'weighted'):
        "The final grade is usually calculated using a **weighted average**. Each component (like exams or homework) has a specific percentage weight. We multiply your score in that component by its weight and sum them up. For example, (Exam Score * 40%) + (Homework Score * 60%).",
    ('improve', 'better', 'score', 'tip', 'boost'):
        "I can certainly help with improvement! My top tips are: **1) Consistent Review:** Don't cram, review notes daily. **2) Practice Problems:** Work through examples until they are easy. **3) Use Office Hours:** Ask your instructor questions immediately when you are confused.",
    ('miss', 'late', 'absent', 'policy', 'extension'):
        "If you miss an assignment or expect to be late, the most important step is to **check your course syllabus immediately** for the stated late work policy. If it's not clear, contact your instructor or TA as soon as possible to discuss a potential solution.",
    ('who are you', 'what are you', 'goodbye', 'bye'):
        "I am the Grade Predictor Assistant Bot, here to help you navigate your course. I hope this was helpful! Good luck with your studies!",
}
def get_response(user_input):
    """
    Processes user input and returns a rule-based response.

    The logic iterates through all defined intent keywords and returns
    the response of the first matching keyword found in the input.
    """
    processed_input = user_input.lower()
    processed_input = re.sub(r'[^\w\s]', '', processed_input)
    for keywords, response in response_data.items():
        if any(keyword in processed_input for keyword in keywords):
            return response
    return "I apologize, I don't have information on that specific topic yet. Please try rephrasing your question, or ask about factors, calculation, or improvement tips!"

print("Chatbot logic successfully defined.")
```

```python
import sys

def run_chatbot():
    """
    Runs the main interactive session for the chatbot.
    """
    print("-" * 55)
    print("Grade Predictor Assistant Chatbot (Type 'quit' or 'exit' to end)")
    print("-" * 55)

    while True:
        try:
            user_input = input("You: ")
            if user_input.lower() in ['quit', 'exit']:
                print("\nBot: Thank you for chatting! Have a productive day.")
                break
            bot_response = get_response(user_input)
            print(f"Bot: {bot_response}")
            print("-" * 55)
        except EOFError:
            # Handles unexpected end-of-file (e.g., if input stream closes)
            print("\n\nSession terminated unexpectedly.")
            break
        except Exception as e:
            # Handles other runtime errors
            print(f"\nAn unexpected error occurred: {e}")
            break

if __name__ == '__main__':
    run_chatbot()
```
